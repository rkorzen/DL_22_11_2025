# Wejście i wyjście danych

Ten notatnik stanowi referencję do operacji wejścia i wyjścia. Pandas potrafi czytać różne typy plików za pomocą metod `pd.read_`. Przyjrzyjmy się najczęstszym typom danych:

```python
import numpy as np
import pandas as pd
```

## CSV
Pliki CSV (Comma Separated Values) to pliki tekstowe, w których pola są rozdzielane przecinkami.<br>
Jeśli nie korzystasz ze środowiska wirtualnego dołączonego do kursu, może być konieczne zainstalowanie <tt>xlrd</tt> i <tt>openpyxl</tt>.<br>
W terminalu/w wierszu poleceń uruchom:

    conda install xlrd
    conda install openpyxl

Następnie zrestartuj Jupyter Notebook.
(lub użyj polecenia pip install, jeśli nie korzystasz z dystrybucji Anaconda)

### Wczytywanie CSV

```python
df = pd.read_csv('example.csv')
df
```

### Zapis do CSV

```python
df.to_csv('example.csv',index=False)
```

## Excel
Pandas potrafi czytać i zapisywać pliki MS Excel. Należy jednak pamiętać, że importowane są tylko dane, a nie formuły czy obrazy. Plik zawierający obrazy lub makra może spowodować błąd metody <tt>.read_excel()</tt>.

### Wczytywanie Excela

```python
pd.read_excel('Excel_Sample.xlsx',sheet_name='Sheet1')
```

### Zapis do Excela

```python
df.to_excel('Excel_Sample.xlsx',sheet_name='Sheet1')
```

## HTML
Pandas potrafi odczytywać tabele z kodu HTML.<br>
Jeśli nie korzystasz ze środowiska wirtualnego dołączonego do kursu, może być konieczna instalacja <tt>lxml</tt>, <tt>html5lib</tt> oraz <tt>BeautifulSoup4</tt>.<br>
W terminalu/w wierszu poleceń uruchom:

    conda install lxml
    conda install html5lib
    conda install beautifulsoup4

Następnie zrestartuj Jupyter Notebook.
(lub użyj polecenia pip install, jeśli nie korzystasz z dystrybucji Anaconda)

### Wczytywanie HTML

Funkcja `pandas.read_html` odczytuje tabele ze stron internetowych i zwraca listę obiektów DataFrame:

```python
df = pd.read_html('http://www.fdic.gov/bank/individual/failed/banklist.html')
```
