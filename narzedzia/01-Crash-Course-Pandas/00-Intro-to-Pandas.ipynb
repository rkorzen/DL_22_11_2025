# Wprowadzenie do Pandas

* Wprowadzenie do Pandas
* Series (Serie)
* DataFrames (Ramki danych)
* Brakujące dane
* GroupBy (Grupowanie)
* Merging, Joining i Concatenating (Łączenie danych)
* Operacje
* Wejście i wyjście danych
