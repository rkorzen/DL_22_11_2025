**ZADANIE: Istnieje kolumna "default". Użyj metody .map() z pandas, aby utworzyć nową kolumnę o nazwie "default code", która zawiera 0, jeśli nie było zaległości, lub 1, jeśli zaległość była. Następnie pokaż head DataFrame z tą nową kolumną.**

[Pomocny link 1](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.map.html)

[Pomocny link 2](https://stackoverflow.com/questions/19798153/difference-between-map-applymap-and-apply-methods-in-pandas)

```python
df['default code'] = df['default'].map({'no':0,'yes':1})
```

```python
df.head()
```

**ZADANIE: Używając metody pandas .apply(), utwórz nową kolumnę o nazwie "marital code". Ta kolumna powinna zawierać skrócony kod, czyli pierwszą literę stanu cywilnego (np. "m" dla "married", "s" dla "single" itd.). Spróbuj zrobić to wyrażeniem lambda. Jest wiele sposobów na to zadanie!**

[Podpowiedź](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.apply.html)


```python
# CODE HERE
```

```python
df['marital code'] = df['marital'].apply(lambda status: status[0])
```

```python
df.head()
```
